```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as skl
```

```python
data=pd.read_csv("training_set.csv")
data
```

```python
#lead_time có NaN là missing value
data["lead_time"] = data["lead_time"].replace(np.NaN, data["lead_time"].median())
```

```python
#perf_6_month_avg có -99 là missing value
perf_6_month_avg=np.array(data['perf_6_month_avg'])
perf_6_month_avg[perf_6_month_avg==-99] = 0.967
print(perf_6_month_avg.shape)
print(perf_6_month_avg)
data['perf_6_month_avg']=perf_6_month_avg
```

```python
#perf_12_month_avg có -99 là missing value
perf_12_month_avg=np.array(data['perf_12_month_avg'])
perf_12_month_avg[perf_12_month_avg==-99] = 0.967
print(perf_12_month_avg.shape)
print(perf_12_month_avg)
data['perf_12_month_avg']=perf_12_month_avg
```

```python
potential_issue = (data['potential_issue'] == 'Yes').astype(float)
deck_risk = (data['deck_risk'] == 'Yes').astype(float)
oe_constraint = (data['oe_constraint'] == 'Yes').astype(float)
ppap_risk = (data['ppap_risk'] == 'Yes').astype(float)
stop_auto_buy = (data['stop_auto_buy'] == 'Yes').astype(float)
rev_stop = (data['rev_stop'] == 'Yes').astype(float)
went_on_backorder = (data['went_on_backorder'] == 'Yes').astype(float)

data['potential_issue'] = potential_issue
data['deck_risk'] = deck_risk
data['oe_constraint'] = oe_constraint
data['ppap_risk'] = ppap_risk
data['stop_auto_buy'] = stop_auto_buy
data['rev_stop'] = rev_stop
data['went_on_backorder'] = went_on_backorder
```

```python
# xóa cột thứ tự đi
del data['sku']
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import time
#Calculating the accuracy and the time taken by the classifier
t0=time.time()
#Data Splicing
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
clf_dt = DecisionTreeClassifier(random_state=0)
#Building the model using the training data set
clf_dt.fit(X_train,y_train)
 
#Evaluating the model using testing data set
y_pred = clf_dt.predict(X_test)
score = accuracy_score(y_test,y_pred)
 
#Printing the accuracy and the time taken by the classifier
print('Accuracy using Decision Tree Classifier:',score) 
print('Time taken using Decision Tree Classifier:' , time.time()-t0) 
```

```python
#Stochastic Gradient Descent Classifier
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import time
t0=time.time()
clf = SGDClassifier(loss='log', random_state=1)
```

```python
import pyprind
pbar = pyprind.ProgBar(45)
for _ in range(45):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
    clf.partial_fit(X_train, y_train, classes=np.unique(y) )
    pbar.update()
print('\nAccuracy: %.10f' % clf.score(X_test, y_test))
```

```python
y_pred = clf.predict(X_test)
score = accuracy_score(y_test,y_pred)
print('Accuracy using Stochastic Gradient Descent Classifier:',score) 
print('Time taken using Stochastic Gradient Descent Classifier:' , time.time()-t0)
```
